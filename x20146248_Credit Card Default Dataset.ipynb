```python
#Importing Required Libraries
import pandas as pd 
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the csv file and putting it into 'df' object.
df = pd.read_csv('UCI_Credit_Card_default.csv')
df.head()
```

```python
#Checking for missing data
df.isnull().sum()
```

```python
# Renaming Columns into more understandable/user-friendly terms

# SEX changed to GENDER
# PAY_0 changed to PAY_1
# default.payment.next.month is too long and changed to something simplier, DEFAULT
df.rename(columns={'SEX':'GENDER',
                   'PAY_0':'PAY_1',
                   'default.payment.next.month':'DEFAULT',} , inplace=True)
```

## Feature Engineering

```python
##Feature Engineering
pay_column = df["PAY_AMT1"] + df["PAY_AMT2"]+ df["PAY_AMT3"]+ df["PAY_AMT4"]+ df["PAY_AMT5"]+ df["PAY_AMT6"]
df['Total_Repayment'] = pay_column

bill_sum = df["BILL_AMT1"]+df["BILL_AMT2"]+df["BILL_AMT3"]+df["BILL_AMT4"]+df["BILL_AMT5"]+df["BILL_AMT6"]
df["Total_bill"]= bill_sum
```

```python
print('Education Column Values: ', df['EDUCATION'].unique())
```

```python
# There exists values 0, 5 and 6 in this column.
# Since these are unknown (undefined), they can be grouped into the category 4: "Others"
def func2(x):
    if(x > 0 and x < 4):
        return x
    else:
        return 4
    
```

```python
df['EDUCATION'] = df['EDUCATION'].apply(func2)
# Changes every value of x not within (and inclusive of) 1 ~ 3 to 4 
```

```python
df['EDUCATION'].unique()
```

```python
df['MARRIAGE'].unique()
```

```python
#There exists value 0 in this cloumn which can be included in others only

def Marriage_convert(x):
    if(x > 0 and x < 3):
        return x
    else:
        return 3
```

```python
df['MARRIAGE'] = df['MARRIAGE'].apply(Marriage_convert)
```

```python
# Corrected changes
df['MARRIAGE'].unique()
```

```python
df['AGE'].unique()
```

```python
# Creating a Function to Distribute the Age
def Age_convert(x):
    if(x>=20 and x<30):
        return 1
    elif(x>=30 and x<40):
        return 2
    elif(x>=40 and x<50):
        return 3
    elif(x>=50 and x<60):
        return 4
    else:
        return 5
```

```python
df['AGE'] = df['AGE'].apply(Age_convert)
```

```python
df['AGE'].unique()
```

```python
# generate binary values using get_dummies
dummy_age = pd.get_dummies(df['AGE'], prefix='AGE' )
dummy_marriage = pd.get_dummies(df['MARRIAGE'], prefix='MARRIAGE' )
dummy_education = pd.get_dummies(df['EDUCATION'],prefix='EDUCATION')
```

```python
print(dummy_age)
```

```python
print(dummy_marriage)
```

```python
print(dummy_education)
```

```python
# merge with main df bridge_df on key values
df = df.join(dummy_age)
df = df.join(dummy_marriage)
df= df.join(dummy_education)
```

```python
df = df.drop(['AGE','MARRIAGE','EDUCATION'],axis=1)
```

```python
df.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
df['Total_Repayment'] = scaling.fit_transform(df[['Total_Repayment']])
df['Total_bill'] = scaling.fit_transform(df[['Total_bill']])
df['LIMIT_BAL'] = scaling.fit_transform(df[['Total_Repayment']])
```

```python

threshold=0.8
```

```python

# find and remove correlated features
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python

correlation(df.iloc[:,:-1],threshold)
```

# Applying Random Forest 

```python
# Importing random forest classifier from sklearn library
from sklearn.ensemble import RandomForestClassifier

# Running the random forest with default parameters.
rfc = RandomForestClassifier()
```

```python
# fit
rfc.fit(X_train,y_train)
```

```python
# Making predictions
predictions = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
# Let's check the report of our default model
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
#As we can infer we get a recall of 0.36 , goal is to maximize the recall score in order to identify maximum defaulters.
from sklearn.metrics import accuracy_score
print("Accuracy {:.2f} %".format( 100 * accuracy_score(y_test,predictions)))

```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(rfc,X,y,cv=10)
score
score.mean()
```
